```python
import numpy as np
import pandas as pd
import time
import os
import datetime as dt
from sklearn.cluster import AgglomerativeClustering
import math
```

```python
#1.Data Prepration


#Please find and download the datasets at "https://bixi.com/en/open-data"


#1.1. 2022 Data
files = os.listdir('2022')
Stations = {'20220104_stations.csv', '20220105_stations.csv','20220106_stations.csv','20220107_stations.csv', '20220108_stations.csv', 
            '202209_stations.csv','202210_stations.csv','202211_stations.csv','202211_deplacements.csv','2022_stations.csv'}

od=files                        
for f in Stations:
        od.remove(f)

origin2022=pd.DataFrame()
des2022=pd.DataFrame()
for f in od:
    print (os.path.join('2022',f))
    d=pd.read_csv(os.path.join('/content/drive/MyDrive/BiXi_Montreal/2022',f))
    d['startst']=d['emplacement_pk_start']
    d['endst']=d['emplacement_pk_end']
    d['date_1']=pd.to_datetime(d['start_date'])
    d['day']=d['date_1'].dt.date
    d['time']=d['date_1'].dt.hour
    d['weekno']=d['date_1'].dt.weekday
    d['weekday']=0
    d['weekend']=0
    d['Mornings']=0
    d['Evenings']=0
    d.loc[d['weekno'] <= 4, 'weekday'] = 1
    d.loc[d['weekno'] > 4, 'weekend'] = 1
    d.loc[ d['time'] >= 7, 'Mornings'] = 1
    d.loc[ d['time'] >= 11, 'Mornings'] = 0
    d.loc[ d['time'] <= 19, 'Evenings'] = 1
    d.loc[ d['time'] <= 15, 'Evenings'] = 0
    d['startst']=d['emplacement_pk_start']
    d['endst']=d['emplacement_pk_end']
    origin=d.pivot_table(index='startst',values=['emplacement_pk_start','day','weekday',	'weekend',	'Mornings',	'Evenings'],
                         aggfunc={'emplacement_pk_start':'count','day':'nunique','weekday':'sum',	'weekend':'sum',	'Mornings':'sum',	'Evenings':'sum'})
    # des=d.pivot_table(index='endst',values=['emplacement_pk_start','day'],aggfunc={'emplacement_pk_start':'count','day':'nunique'})
    origin2022= origin2022.append(origin)
    # des2022= des2022.append(des)

o2022=origin2022.pivot_table(index='startst',values=['emplacement_pk_start','day','weekday',	'weekend',	'Mornings',	'Evenings'],aggfunc='sum')   #Pivot_Table_Total Trips/days/stations
# d2022=des2022.pivot_table(index='endst',values=['emplacement_pk_start','day'],aggfunc='sum')
o2022['trips_count']=o2022['emplacement_pk_start']
o2022['tot_trip']=o2022['trips_count'].sum()
o2022['Evening_%']=100* ( o2022['Evenings']/o2022['trips_count'])
o2022['Morning_%']=100* ( o2022['Mornings']/o2022['trips_count'])
o2022['weekday_%']=100* ( o2022['weekday']/o2022['trips_count'])
o2022['weekend_%']=100* ( o2022['weekend']/o2022['trips_count'])
o2022['trip_per_day']=o2022['trips_count'].sum()
tot2022=o2022['tot_trip'].mean()
weekdays=o2022['weekday'].sum()
weekdays_perc= o2022['weekday_%'].mean()
weekends=o2022['weekend'].sum()
weekends_perc= o2022['weekend_%'].mean()

Mornings=o2022['Mornings'].sum()
Mornings_perc= o2022['Morning_%'].mean()
Evenings=o2022['Evenings'].sum()
Evenings_perc= o2022['Evening_%'].mean()

print('total trips 2022:', tot2022)
print('weekdays 2022:', weekdays)
print('weekdays_perc 2022:', weekdays_perc)
print('weekends 2022:', weekends)
print('weekends_perc 2022:', weekends_perc)
print('Mornings trips 2022:', Mornings)
print('Mornings_perc trips 2022:', Mornings_perc)
print('Evenings trips 2022:', Evenings)
print('Evenings_perc 2022:', Evenings_perc)
```
